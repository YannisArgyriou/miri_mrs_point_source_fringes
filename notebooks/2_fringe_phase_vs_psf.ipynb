### Preamble

### --Fitting functions

```python
def polyfit2d(x, y, z, order=3):
    ncols = (order + 1)**2
    G = np.zeros((x.size, ncols))
    ij = itertools.product(range(order+1), range(order+1))
    for k, (i,j) in enumerate(ij):
        G[:,k] = x**j * y**i
    m, _, _, _ = np.linalg.lstsq(G, z)
    return m

def polyval2d(x, y, m):
    order = int(np.sqrt(len(m))) - 1
    ij = itertools.product(range(order+1), range(order+1))
    z = np.zeros_like(x)
    for a, (i,j) in zip(m, ij):
        z = z + a * x**j * y**i
    return z
```

### --Analysis
* Compute the fringe phase as a function of distance from the PSF peak 
* Perform a 2D polynomial fit as a function of distance from PSF peak and wavelength (the incidence angle is slightly different at different parts of the detector in the wavelength dimension)

```python
centered_alpha_wvnrs = {}
ps_wvnr_lists = {}
ps_fringe_phases = {}
fringe_phase_polyfit_weights = np.zeros((3,(polyfit_order + 1)**2))

for slice_offset in [-1,0,1]:
    print('Slice offset {}'.format(slice_offset))
    print('---------------')
    x,y,z = [],[],[]
    
    ypos,xpos = detpixel_trace_compactsource(sci_img,band,d2cMaps,offset_slice=slice_offset)
    xpos = (np.ones(len(ypos))*xpos[512]).astype(int)
    
    for xpos_offset in [-4,-3,-2,-1,0,1,2]:
        print('XPOS_OFFSET={}'.format(xpos_offset))

        wavl = d2cMaps['lambdaMap'][ypos,xpos+xpos_offset]
        extracted_spectrum =  sci_img[ypos,xpos+xpos_offset]

        sel = np.isfinite(extracted_spectrum)
        wavl = wavl[sel]
        extracted_spectrum = extracted_spectrum[sel]

        positive_peaks_idxs = find_peaks(extracted_spectrum,thres=thres,min_dist=min_dist)

        # Scale to 1d spectrum continuum level (defined by fringe peaks)
        arr_interpolator = scp_interpolate.InterpolatedUnivariateSpline(positive_peaks_idxs,extracted_spectrum[positive_peaks_idxs],k=2,ext=3)
        arr_profile = arr_interpolator(range(len(extracted_spectrum)))

        fringe_transmission = extracted_spectrum/arr_profile
        fringe_transmission[(fringe_transmission >= 1.)] = 1

        # Point source fringe analysis
        ps_wvnr = np.flipud(10000./wavl)
        ps_spec = np.flipud(fringe_transmission)-0.005
        ps_peaks_wvnr = np.flipud(len(ps_wvnr)-1-np.array(positive_peaks_idxs))

        # point source placeholders
        ps_wvnr_list = ps_wvnr[ps_peaks_wvnr]
        ps_R_list = np.full(len(ps_peaks_wvnr),np.nan)
        ps_D_list = np.full(len(ps_peaks_wvnr),np.nan)

        valid_range = np.arange(2,len(ps_peaks_wvnr)-2)
        for i in valid_range:
            if i == valid_range[0]:
                R,D = R0,D0
            else:
                R,D = ps_R_list[i-1],ps_D_list[i-1]
            if np.isfinite(R) == False:
                R,D = R0,D0
            if (np.flipud(extracted_spectrum)[ps_peaks_wvnr[i]] <min_signal_to_fit):
                ps_R_list[i] = np.nan
                ps_D_list[i] = np.nan
                continue
            else:
                sel = (ps_wvnr>(ps_wvnr[ps_peaks_wvnr[i-1]]+ps_wvnr[ps_peaks_wvnr[i]])/2. ) & (ps_wvnr<(ps_wvnr[ps_peaks_wvnr[i]]+ps_wvnr[ps_peaks_wvnr[i+1]])/2. )
                try:
                    popt,pcov = curve_fit(FPfunc_noPhaseShift,ps_wvnr[sel],ps_spec[sel],p0=[R,D],bounds=((0,0),(1,np.inf)) )
                    ps_R_list[i] = popt[0]
                    ps_D_list[i] = popt[1]
                except:
                    print("Error fitting peak {}".format(i)+str(IOError))
                    ps_R_list[i] = ps_R_list[i-1]
                    ps_D_list[i] = ps_D_list[i-1]

                # compute linecenter empirically
                plotx = np.linspace(ps_wvnr[sel][0],ps_wvnr[sel][-1],10000)
                ploty = FPfunc_noPhaseShift(plotx,ps_R_list[i],ps_D_list[i])
                ps_wvnr_list[i] = plotx[np.argmax(ploty)]
        valid_range = np.where(np.isfinite(ps_D_list)==True)[0]

        # Extended source fringe analysis
        ext_peaks_wvl =  find_peaks(fringe_flat[ypos,xpos+xpos_offset],thres=thres,min_dist=min_dist)
        if (len(np.arange(3,len(ext_peaks_wvl)-2))<len(valid_range)) & (xpos_offset>0):
            # if the column is too close to the edge of a slice, the fringe_flat is full of ones, yielding less fringes. We mitigate this by selecting a column to the left as reference instead.
            flag = 1
            xpos_offset -= 1
            ext_peaks_wvl =  find_peaks(fringe_flat[ypos,xpos+xpos_offset],thres=thres,min_dist=min_dist)
        else:
            flag = 0
        ext_fringe =  fringe_flat[ypos,xpos+xpos_offset] # ext_source_norm[0]/ext_source_norm[2]

        ext_wvnr = np.flipud(10000./d2cMaps['lambdaMap'][ypos,xpos+xpos_offset])
        ext_spec = np.flipud(ext_fringe)
        ext_peaks_wvnr = np.flipud(len(ext_wvnr)-1-ext_peaks_wvl)

        # placeholders
        ext_wvnr_list = ext_wvnr[ext_peaks_wvnr]
        ext_R_list = np.full(len(ext_peaks_wvnr),np.nan)
        ext_D_list = np.full(len(ext_peaks_wvnr),np.nan)

        for i in range(3,len(ext_peaks_wvnr)-2):
            if i == 3:
                R,D = R0,D0
            else:
                R = ext_R_list[i-1]
                D = ext_D_list[i-1]
            if np.isfinite(R) == False:
                R,D = R0,D0
            sel = (ext_wvnr>(ext_wvnr[ext_peaks_wvnr[i-1]]+ext_wvnr[ext_peaks_wvnr[i]])/2. ) & (ext_wvnr<(ext_wvnr[ext_peaks_wvnr[i]]+ext_wvnr[ext_peaks_wvnr[i+1]])/2. )

            try:
                popt,pcov = curve_fit(FPfunc_noPhaseShift,ext_wvnr[sel],ext_spec[sel],p0=[R,D],bounds=((0,0),(1,np.inf)) )
                ext_R_list[i] = popt[0]
                ext_D_list[i] = popt[1]
            except:
                ext_R_list[i] = np.nan
                ext_D_list[i] = np.nan

            # compute linecenter empirically
            plotx = np.linspace(ext_wvnr[sel][0],ext_wvnr[sel][-1],10000)
            ploty = FPfunc_noPhaseShift(plotx,ext_R_list[i],ext_D_list[i])
            ext_wvnr_list[i] = plotx[np.argmax(ploty)]
        if flag == 1:
            xpos_offset += 1

        # Distance from PSF peak
        interp_alpha = scp_interpolate.interp1d(np.flipud(10000./d2cMaps['lambdaMap'][ypos,xpos+xpos_offset]),np.flipud(d2cMaps['alphaMap'][ypos,xpos+xpos_offset]) )
        alpha_wvnr = interp_alpha(ps_wvnr_list)
        centered_alpha_wvnr = alpha_wvnr-poly_alpha(10000./ps_wvnr_list)

        # Compute point source phase shift
        ps_delta_phi = np.full(len(ps_wvnr_list),np.nan)
        for i in valid_range:
            closest_idx = find_nearest(ext_wvnr[ext_peaks_wvnr],ps_wvnr[ps_peaks_wvnr[i]])

            sel = (ext_wvnr>(ext_wvnr[ext_peaks_wvnr[closest_idx-1]]+ext_wvnr[ext_peaks_wvnr[closest_idx]])/2. ) & (ext_wvnr<(ext_wvnr[ext_peaks_wvnr[closest_idx]]+ext_wvnr[ext_peaks_wvnr[closest_idx+1]])/2. )
            test_wvnr = np.arange(ext_wvnr[sel][0],ext_wvnr[sel][-1],0.01)
            ps_delta_phi[i] = 2*np.pi*ext_D_list[closest_idx] * (ext_wvnr_list[closest_idx]-ps_wvnr_list[i])

            if (ps_delta_phi[i] <-2.5) or (ps_delta_phi[i] > 1.5):
                closest_idx -= 1

                sel = (ext_wvnr>(ext_wvnr[ext_peaks_wvnr[closest_idx-1]]+ext_wvnr[ext_peaks_wvnr[closest_idx]])/2. ) & (ext_wvnr<(ext_wvnr[ext_peaks_wvnr[closest_idx]]+ext_wvnr[ext_peaks_wvnr[closest_idx+1]])/2. )
                test_wvnr = np.arange(ext_wvnr[sel][0],ext_wvnr[sel][-1],0.01)
                ps_delta_phi[i] = 2*np.pi*ext_D_list[closest_idx] * (ext_wvnr_list[closest_idx]-ps_wvnr_list[i])

        sel = np.isfinite(ps_delta_phi)
        centered_alpha_wvnr = centered_alpha_wvnr[sel]
        ps_wvnr_list = ps_wvnr_list[sel]
        ps_delta_phi = ps_delta_phi[sel]

        x.extend(centered_alpha_wvnr)
        y.extend(ps_wvnr_list)
        z.extend(ps_delta_phi)

    centered_alpha_wvnrs['slice_offset_{}'.format(slice_offset)] = np.array(x)
    ps_wvnr_lists['slice_offset_{}'.format(slice_offset)] = np.array(y)
    ps_fringe_phases['slice_offset_{}'.format(slice_offset)] = np.array(z)

    print('')

    # Fit 2d polynomial to data
    sel = (ps_fringe_phases['slice_offset_{}'.format(slice_offset)]>-2) & (ps_fringe_phases['slice_offset_{}'.format(slice_offset)]<0.9)
    fringe_phase_polyfit_weights[slice_offset+1,:] = polyfit2d(centered_alpha_wvnrs['slice_offset_{}'.format(slice_offset)][sel],ps_wvnr_lists['slice_offset_{}'.format(slice_offset)][sel],ps_fringe_phases['slice_offset_{}'.format(slice_offset)][sel],polyfit_order)
```

### --Plot raw results in one detector slice

```python
slice_offset = 0
key = 'slice_offset_'+str(slice_offset)

plt.close('all')
plt.figure(figsize=(12,4))
plt.scatter(centered_alpha_wvnrs[key],ps_fringe_phases[key],c=ps_wvnr_lists[key])
plt.ylim(-2,1)
plt.xlabel('Distance from PSF peak [arcsec]')
plt.ylabel('Fringe phase-shift [rad]')
plt.title('CV2_P0 // Slice offset '+str(slice_offset))
plt.tight_layout()

fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')
ax.scatter(centered_alpha_wvnrs[key],ps_wvnr_lists[key],ps_fringe_phases[key],c=ps_wvnr_lists[key],alpha=1)
ax.set_zlim(-2,1.01)
ax.set_xlabel(r'Distance from PSF peak [arcsec]')
ax.set_ylabel(r'Wavenumber [cm$^{-1}$]')
ax.set_zlabel(r'Fringe phase-shift [rad]')
ax.set_title('CV2_P0 // Slice offset '+str(slice_offset))
ax.tick_params(axis='both',labelsize=10)
plt.tight_layout()
```

### --Plot 2D fit results

```python
# 2D polyfit weights
m = fringe_phase_polyfit_weights[slice_offset+1,:]

# Evaluate the 2D polynomial fit on a regular grid...
nx, ny = 20, 20
xx, yy = np.meshgrid(np.linspace(centered_alpha_wvnrs[key].min(), centered_alpha_wvnrs[key].max(), nx),
                     np.linspace(ps_wvnr_lists[key].min(), ps_wvnr_lists[key].max(), ny))
zz = polyval2d(xx, yy, m)
pred = polyval2d(centered_alpha_wvnrs[key], ps_wvnr_lists[key], m)

plt.close('all')
fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')
ax.scatter(centered_alpha_wvnrs[key],ps_wvnr_lists[key],ps_fringe_phases[key],c=ps_wvnr_lists[key],alpha=1)
ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, alpha=0.2)
ax.set_zlim(-2,1)
ax.set_xlabel(r'Distance from PSF peak [arcsec]')
ax.set_ylabel(r'Wavenumber [cm$^{-1}$]')
ax.set_zlabel(r'Fringe phase-shift [rad]')
ax.tick_params(axis='both',labelsize=10)
ax.set_title('2D fit')
plt.tight_layout()

sel = (ps_fringe_phases[key] > -2.) & (ps_fringe_phases[key] < 0.9)

fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')
ax.scatter(centered_alpha_wvnrs[key][sel],ps_wvnr_lists[key][sel],(ps_fringe_phases[key]-pred)[sel],c=ps_wvnr_lists[key][sel],alpha=1)
ax.set_xlabel(r'Distance from PSF peak [arcsec]')
ax.set_ylabel(r'Wavenumber [cm$^{-1}$]')
ax.set_zlabel(r'Residuals')
ax.tick_params(axis='both',labelsize=10)
ax.set_title('2D fit residuals')
plt.tight_layout()

plt.figure(figsize=(9,4))
(mu, sigma) = norm.fit((ps_fringe_phases[key]-pred)[sel][abs(ps_fringe_phases[key]-pred)[sel]<0.2])
n, bins, patches = plt.hist((ps_fringe_phases[key]-pred)[sel][abs(ps_fringe_phases[key]-pred)[sel]<0.2], 20, density=1, alpha=0.5,label=r'$\mu$ = {}, $\sigma$ = {}'.format(round(mu,3),round(sigma,3)))
y = norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--', linewidth=2)
plt.xlabel('Residuals')
plt.ylabel('Probability')
plt.title('Histogram of residuals')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
