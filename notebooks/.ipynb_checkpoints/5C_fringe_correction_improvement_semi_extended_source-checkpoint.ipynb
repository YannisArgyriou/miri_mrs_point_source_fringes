### --Load data

```python
ps_img        = fits.open(datapath+'CV2_P0.fits')[1].data
ext_img       = fits.open(datapath+'FM_800K_BB_EXTENDED.fits')[1].data
fringe_flat   = np.load(datapath+'fringe_flat.npy')
spectral_response = np.load(datapath+'spectral_response.npy')
```

### --Apply spectral response correction

```python
ps_img_photom  =  ps_img/spectral_response
ext_img_photom = ext_img/spectral_response
```

### --Load point source fringe flat from notebook 4

```python
ps_fringe_flat = np.load(outputpath+'ps_fringe_flat.npy')
```

### Simulate field

### --Point source on top of semi-extended source
* We convolve the point source image with a 6 pixel wide box kernel. This represents the semi-extended component.
* We multiply the point source signal by a constant factor to better represent the flux ratio between a bright point source surrounded by a disc of cooler temperature.
* Limit for computation of standard deviation on residuals set to 5 DN/s.

```python
semi_ext_source = np.zeros((1024,1032))
semi_ext_source_ps_fringe_flat = np.zeros((1024,1032))
for row in range(1024):
    semi_ext_source[row,:516] = 2*convolve(ps_img_photom[row,:516],Box1DKernel(6))
    semi_ext_source_ps_fringe_flat[row,:516] = 2*convolve((ps_img_photom/ps_fringe_flat)[row,:516],Box1DKernel(6))
```

```python
point_source    = 5*ps_img_photom

semi_ext_source_plus_point_source  = semi_ext_source + point_source
semi_ext_source_plus_point_source_fringe_flat   = semi_ext_source_plus_point_source/fringe_flat
semi_ext_source_plus_point_source_forward_model = semi_ext_source_ps_fringe_flat + (point_source/ps_fringe_flat)

fig, (axs0, axs1) = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'wspace':0,'width_ratios': [1, 3]})
axs0.plot(semi_ext_source[512,22:36],np.arange(22,36),label='Disc',alpha=0.4)
axs0.plot(point_source[512,22:36],np.arange(22,36),label='Point source',alpha=0.4)
axs0.plot(semi_ext_source_plus_point_source[512,22:36],np.arange(22,36),label='Disc+Point source' )
axs0.hlines(np.arange(24,32),0,200,alpha=0.2)
axs0.set_xlim(0,150)
axs0.set_ylim(23,32.)
axs0.set_xlabel('Signal [mJy]',fontsize=20)
axs0.set_ylabel('Detector column',fontsize=20)
axs0.tick_params(axis='both',labelsize=16)
axs0.legend(framealpha=1,fontsize=12)
axs0.invert_xaxis()
for col in range(24,32):
#     deal with NaNs and bad pixels
    if col == 24:
        sel = (semi_ext_source_plus_point_source_forward_model[:,col]>5) & np.isfinite(semi_ext_source_plus_point_source_forward_model[:,col])  & ~((np.arange(1024)>647) & (np.arange(1024)<653))
    else:
        sel = (semi_ext_source_plus_point_source_forward_model[:,col]>5) & np.isfinite(semi_ext_source_plus_point_source_forward_model[:,col])
    
    residuals_fringe_flat    = (((semi_ext_source_plus_point_source_fringe_flat)[:,col][sel]/gaussian_filter((semi_ext_source_plus_point_source_fringe_flat)[:,col][sel],5))-1)
    residuals_ps_fringe_flat = ((semi_ext_source_plus_point_source_forward_model[:,col][sel]/gaussian_filter(semi_ext_source_plus_point_source_forward_model[:,col][sel],5))-1)
    
    axs1.plot(np.arange(1024)[sel],col+2.5*residuals_fringe_flat,'k')    # multiply residuals by factor 2.5 for easier visualization
    axs1.plot(np.arange(1024)[sel],col+2.5*residuals_ps_fringe_flat,'r') # multiply residuals by factor 2.5 for easier visualization
    
    
    axs1.text(430, col+0.4, r'1$\sigma$ = {}%   vs'.format(round(100*np.std(residuals_fringe_flat),1)),color='k')
    axs1.text(580, col+0.4, r'{}%'.format(round(100*np.std(residuals_ps_fringe_flat),1)),color='r')
axs1.hlines(np.arange(24,32),0,1024,alpha=0.2)
legend_elements = [Line2D([0],[0],color='k',label='fringe flat'),
                   Line2D([0],[0],color='r',label='forward model')]
axs1.legend(handles=legend_elements,loc='upper left',framealpha=1,fontsize=12)
axs1.set_xlim(0,1024)
axs1.set_ylim(23,32.)
axs1.set_yticks([])
axs1.set_xlabel('Detector row',fontsize=20)
axs1.tick_params(axis='both',labelsize=16)
fig.tight_layout()
if save_plots:
    plt.savefig(figurepath+'point_source_and_disc_residuals.pdf',dpi=2540)
```
